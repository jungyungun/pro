```python
from selenium import webdriver
from bs4 import BeautifulSoup
from openpyxl import Workbook
import openpyxl
import re
import datetime
import os
import shutil

#======== 셀 레 니 움 설 정 ========#
options = webdriver.ChromeOptions()
#헤들리스로 만들기-----------------

#options.add_argument('headless')
#options.add_argument('--disable-gpu')
#options.add_argument('lang=ko_KR')
#----------------------------------
path = 'C:\chromedriver.exe'
driver = webdriver.Chrome(path, chrome_options=options)
#======== 셀 레 니 움 설 정 ========#


#폴더 생성을 위해 오늘의 날짜를 받아옴
TODAY_DATE = datetime.datetime.today()

#키워드 생성 알고리즘#
Keyword_list = ['발전사', '한전', '5대발전','전력','태양광발전','신재생에너지','석유','석탄','전기세']
report_list=['에너지경제연구원','포스코경영연구원','LG경제연구원','한전경제연구원','현대경제연구원','수출입은행해외경제연구소','코트라','삼성경제연구소']

#키워드의 엑셀파일을 만드는 알고리즘#    
i = 0
for i in range(len(Keyword_list)):
    if os.path.isfile('C:/Users/wonjo/Database/'+Keyword_list[i]+'.xlsx') == False:
        wb = Workbook()
        wb.save('C:/Users/wonjo/Database/'+Keyword_list[i]+'.xlsx')
        TODAY_DATE = datetime.datetime.today()
i = 0    
for i in range(len(report_list)):
    if os.path.isfile('C:/Users/wonjo/Database/'+report_list[i]+'.xlsx')==False:
        wb = Workbook()
        wb.save('C:/Users/wonjo/Database/'+report_list[i]+'.xlsx')
```

```python
#### 뉴스 크롤러 ####

import openpyxl
def news_crowler(Keyword_list) : 
    i = 0 ## 다음 키워드로 넘겨주는 변수

    for i in range(len(Keyword_list)): #리스트의 키워드를 반복하기 위한 For 반복문#
        #키워드의 엑셀파일을 열고, 
        wb = openpyxl.load_workbook('C:/Users/wonjo/Database/'+Keyword_list[i]+'.xlsx')
        wb.create_sheet(TODAY_DATE.strftime('%Y-%m-%d'))
        Today_Sheet = wb[TODAY_DATE.strftime('%Y-%m-%d')]

        link_partition_1 = 'https://www.google.com/search?q='
        link_partition_2 = Keyword_list[i]
        link_partition_3 = '+-스포츠%2C+-손흥민%2C+-농구%2C+-전쟁%2C+-야구%2C+-리버풀%2C+-축구%2C+-軍%2C+-핵무기%2C+-공포%2C+-감독&oq=전력+-스포츠%2C+-손흥민%2C+-농구%2C+-전쟁%2C+-軍%2C+-핵무기%2C+-공포%2C+-감독&lr=lang_ko%7Clang_zh-CN%7Clang_vi&hl=ko&tbas=0&tbs=lr:lang_1ko%7Clang_1zh-CN%7Clang_1vi,qdr:d&tbm=nws&source=lnt&sa=X&ved=0ahUKEwjtjYCNsaHgAhVC6LwKHQJSAG0QpwUIHQ&biw=1536&bih=754&dpr=1.25'
        #link_partition_3은 국내 뉴스만 검색하게끔 만듦.
        Current_URL = link_partition_1 + link_partition_2 + link_partition_3 #링크를 합치는 구조임#
        
        crowling_count = 0 #크롤링을 한 횟수를 카운트하기 위해 만들어주는 변수 매번 0으로 갱신
        row = 0 #엑셀 파일을 이어붙이기위해 매번 row를 0으로 갱신
        row_index = 0
        while (1):
            driver.get(Current_URL)
            html = driver.page_source    
            soup = BeautifulSoup(html,'html.parser')  
            driver.implicitly_wait(111)
            
            #크롤링한 데이터를 엑셀에 옮겨적는 알고리즘
            for row_index in range(len(soup.select("h3 > a"))):
                Today_Sheet.cell(row=row_index+1 + crowling_count, column=1).value = soup.select("h3 > a")[row_index].text
                Today_Sheet.cell(row=row_index+1 + crowling_count, column=2).value = soup.select("h3 > a")[row_index].get('href')
                Today_Sheet.cell(row=row_index+1 + crowling_count, column=3).value = TODAY_DATE.strftime('%Y-%m-%d')
            #페이지 넘기기 알고리즘
            if len(soup.select('td > a.pn')) == 2 :
                next_page = soup.select('a.pn')[1].get('href')
            elif len(soup.select('td > a.pn')) == 1 :
                if'다음' in soup.select('td > a.pn')[0].text :
                    next_page = soup.select('a.pn')[0].get('href')
                else: 
                    break
            elif len(soup.select('td > a.pn')) == 0 :
                break
            #다음페이지로 이동하는 코드 
            Current_URL = Current_URL+next_page
            
            #다음페이지로 넘길 때, 크롤링카운트를 10증가시킴
            crowling_count = crowling_count + 10 
            
        #데이터의 저장
        wb.save('C:/Users/wonjo/Database/'+Keyword_list[i]+'.xlsx')
#명령문 입력
news_crowler(Keyword_list)

    
```

```python
#### 뉴스 업로더 ####

import openpyxl
from openpyxl import Workbook
from selenium.webdriver.support.ui import Select
#로그인되었으면 다음을 주석처리#

def news_uploader(Keyword_list) : 
    
    driver.get('http://127.0.0.1:8000/admin/MorningBriefing/news/')
    html = driver.page_source    
    soup = BeautifulSoup(html,'html.parser')

    if soup.select('h1 > a')[0].text == 'Django administration':
        driver.implicitly_wait(100)
        Username = driver.find_element_by_id('id_username')
        Password = driver.find_element_by_name('password')
        Username.send_keys('admin')
        Password.send_keys('admin')
        Password.submit()
    i = 0 ## 다음 키워드로 넘겨주는 변수
    for i in range(len(Keyword_list)): #리스트의 키워드를 반복하기 위한 For 반복문#
        #키워드의 엑셀파일을 열고, 데이터를 불러옴
        wb = openpyxl.load_workbook('C:/Users/wonjo/Database/'+Keyword_list[i]+'.xlsx')
        Today_Sheet = wb[TODAY_DATE.strftime('%Y-%m-%d')]
        j = 0
        if Today_Sheet.max_column == 3:
            #엑셀의 데이터를 장고로 옮겨주는 알고리즘
            for j in range(Today_Sheet.max_row):
                driver.get('http://127.0.0.1:8000/admin/MorningBriefing/news/add/')
                html = driver.page_source    
                soup = BeautifulSoup(html,'html.parser')
                driver.implicitly_wait(100)
            
            # 제목/ 링크/ 업로드 버튼을 따옴#
                title = driver.find_element_by_id('id_Title')
                link = driver.find_element_by_id('id_Link')
                data_field = Select(driver.find_element_by_id('id_Data_field'))
                write =driver.find_element_by_name('_addanother')
                driver.implicitly_wait(100)
                title.send_keys(Today_Sheet.cell(row=j+1, column=1).value)
                driver.implicitly_wait(100)
                link.send_keys(Today_Sheet.cell(row=j+1, column=2).value)
                driver.implicitly_wait(100)
                
                if Keyword_list[i] == '발전사' or Keyword_list[i] ==  '한전' or Keyword_list[i] ==  '5대발전':
                    data_field.select_by_value('A')
                if Keyword_list[i] == '전력':
                    data_field.select_by_value('B')
                if Keyword_list[i] == '태양광발전' or Keyword_list[i] == '신재생에너지':
                    data_field.select_by_value('C')
                if Keyword_list[i] == '석유' or Keyword_list[i] ==  '천연가스' or Keyword_list[i] ==  '석탄'or Keyword_list[i] ==  '전기세':
                    data_field.select_by_value('D')
                write.submit()
                driver.implicitly_wait(100)
        wb.close()
#명령문입력#
news_uploader(Keyword_list)
```
